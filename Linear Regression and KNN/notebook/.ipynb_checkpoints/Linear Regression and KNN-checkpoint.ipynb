```python
def remove_outlier(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1   #IQR is interquartile range. 

    filter_col = (data[col] >= Q1 - 1.5 * IQR) & (data[col] <= Q3 + 1.5 *IQR)
    return data.loc[filter_col]  
```

```python
# create new data subset with removing outlier
data_filter_Outlier = remove_outlier(data, 'Ambient_Pressure')
data_filter_Outlier = remove_outlier(data_filter_Outlier, 'Relative_Humidity')
```

#### (h) Can you improve your model using possible interaction terms or nonlinear associations between the predictors and response? Train the regression model on a randomly selected 70% subset of the data with all predictors. Also, run a regression model involving all possible interaction terms and quadratic nonlinearities, and remove insignificant variables using p-values (be careful about interaction terms). Test both models on the remaining points and report your train and test MSEs.

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(data, test_size=0.3)
x_train = train.iloc[:, :-1]
y_train = train.iloc[:, -1]
x_test = test.iloc[:, :-1]
y_test = test.iloc[:, -1]
```

```python
from sklearn import metrics

# Original Linear Regression

X_train = sm.add_constant(x_train)
linear_regressor = LinearRegression().fit(X_train, y_train)
original_train_prediction = linear_regressor.predict(X_train)


X_test = sm.add_constant(x_test)
original_test_prediction = linear_regressor.predict(X_test)

print("============ Train Error And Test Error Before Any Modification ============")
print("train error: " + str(round(metrics.mean_squared_error(y_train, original_train_prediction),3)))
print("test error: " + str(round(metrics.mean_squared_error(y_test, original_test_prediction),3)))
```

```python
labels = ['const','Ambient_Temp', 'Exhaust_Vacuum', 'Ambient_Pressure',  'Relative_Humidity', 
                      'Ambient_Temp x Exhaust_Vacuum', 'Ambient_Temp x Ambient_Pressure',
                      'Ambient_Temp x Relative_Humidity', 'Exhaust_Vacuum x Ambient_Pressure',
                      'Exhaust_Vacuum x Relative_Humidity', 'Ambient Pressure x Relative Humidity',
                      'Ambient_Temp^2', 'Exhaust Vacuum^2', 'Ambient_Pressure^2', 'Relative_Humidity^2']

poly = PolynomialFeatures(degree=2)
x_train_poly = poly.fit_transform(x_train)
x_train_poly = sm.add_constant(x_train_poly)
x_test_poly = poly.fit_transform(x_test)
x_test_poly = sm.add_constant(x_test_poly)

linear_regressor = LinearRegression().fit(x_train_poly, y_train)

interaction_train_prediction = linear_regressor.predict(x_train_poly)
interaction_test_prediction = linear_regressor.predict(x_test_poly)

print("============ Train Error And Test Error After Adding Interactions ============")
print("train error: " + str(round(metrics.mean_squared_error(y_train, interaction_train_prediction),3)))
print("test error: " + str(round(metrics.mean_squared_error(y_test, interaction_test_prediction),3)))

model = sm.OLS(y_train, x_train_poly)
inter_results = model.fit()
print(inter_results.summary(xname = labels))
```

```python
# Filter the parameter, removing the indicators with p-value > 0.05
remaining_index = []
for i in range(len(inter_results.pvalues)):
    if inter_results.pvalues.index[i] in list(inter_results.pvalues[inter_results.pvalues < 0.05].index):
        remaining_index.append(i)

x_train_remaining = []
x_test_remaining = []

for data in x_train_poly:
    x_train_remaining.append([data[i] for i in remaining_index]) 
    
for data in x_test_poly:
    x_test_remaining.append([data[i] for i in remaining_index])  
    
x_train_remaining = np.array(x_train_remaining)

x_test_remaining = np.array(x_test_remaining)

linear_regressor = LinearRegression().fit(x_train_remaining, y_train)

interaction_rm_train_prediction = linear_regressor.predict(x_train_remaining)
interaction_rm_test_prediction = linear_regressor.predict(x_test_remaining)

print("============ Train Error And Test Error After Adding Interactions(Removing Insignificant Variables) ============")
print("train error: " + str(round(metrics.mean_squared_error(y_train, interaction_rm_train_prediction),3)))
print("test error: " + str(round(metrics.mean_squared_error(y_test, interaction_rm_test_prediction),3)))

model = sm.OLS(y_train, x_train_remaining)
inter_rm_results = model.fit()
print(inter_rm_results.summary())
```

Findings:  
By adding polynomial features, training and testing MSE are decreased.  
However, I do not understand why the MSE increase a little bit after I remove some insignificant indicator.

#### (i) KNN Regression: Perform k-nearest neighbor regression for this dataset using both normalized and raw features. Find the value of k  {1,2,...,100} that gives you the best fit. Plot the train and test errors in terms of 1/k.

```python
from sklearn.preprocessing import Normalizer
from sklearn.neighbors import KNeighborsRegressor

data = pd.read_excel('../data/CCPP/Folds5x2_pp.xlsx', sheet_name = 'Sheet1')
data.rename(columns={"AT": "Ambient_Temp", "V": "Exhaust_Vacuum", "AP": "Ambient_Pressure", "RH": "Relative_Humidity", "PE": "Energy_Output"}, inplace=True)

#x_data = data.iloc[:, :-1]
#y_data = data['Energy_Output']

transformer = Normalizer().fit(data)
normalize_data = transformer.transform(data)

normalize_train, normalize_test = train_test_split(normalize_data, test_size=0.3)
normalize_x_train = normalize_train[:, :-1]
normalize_y_train = normalize_train[:, -1]
normalize_x_test = normalize_test[:, :-1]
normalize_y_test = normalize_test[:, -1]

train, test = train_test_split(data, test_size=0.3)
x_train = train.iloc[:, :-1]
y_train = train.iloc[:, -1]
x_test = test.iloc[:, :-1]
y_test = test.iloc[:, -1]

train_mse, test_mse = {}, {}
train_normalized_mse, test_normalized_mse = {}, {}

for i in range(1,101):
    knn_normalize = KNeighborsRegressor(n_neighbors=i).fit(normalize_x_train, normalize_y_train)
    knn_raw = KNeighborsRegressor(n_neighbors=i).fit(x_train, y_train)
    
    # predict the training data and testint data of "normalize_data"
    y_normalize_train_predict = knn_normalize.predict(normalize_x_train)
    y_normalize_test_predict = knn_normalize.predict(normalize_x_test)
    
    # predict the training data and testint data of "raw_data"
    y_train_raw_predict = knn_raw.predict(x_train)
    y_test_raw_predict = knn_raw.predict(x_test)
    
    # store the error rate
    train_normalized_mse[i] = metrics.mean_squared_error(normalize_y_train, y_normalize_train_predict)
    test_normalized_mse[i] = metrics.mean_squared_error(normalize_y_test, y_normalize_test_predict)
    train_mse[i] = metrics.mean_squared_error(y_train, y_train_raw_predict)
    test_mse[i] = metrics.mean_squared_error(y_test, y_test_raw_predict)
    
```

```python
k_range = np.arange(1, 101)

plt.title('KNN Error Rate with number of neighbors(Raw)')

plt.plot(1 / k_range, train_normalized_mse.values(), label = 'Train Error Rate(Normalize)')
plt.plot(1 / k_range, test_normalized_mse.values(), label = 'Test Error Rate(Normalize)')

plt.legend()
#plt.grid(True)
plt.xlabel('1 / Number of neighbors(k)')
plt.ylabel('Mean Squared Error')
plt.show()
```

```python
k_range = np.arange(1, 101)

plt.title('KNN Error Rate with number of neighbors(Normalize)')

plt.plot(1 / k_range, train_mse.values(), label = 'Train Error Rate(Raw)')
plt.plot(1 / k_range, test_mse.values(), label = 'Test Error Rate(Raw)')

plt.legend()
#plt.grid(True)
plt.xlabel('1 / Number of neighbors(k)')
plt.ylabel('Mean Squared Error')
plt.show()
```

### ISLR: 2.4.1  
For each of parts (a) through (d), indicate whether we would generally expect the performance of a flexible statistical learning method to be better or worse than an inflexible method. Justify your answer.

#### (a) The sample size n is extremely large, and the number of predictors p is small.

Better. The flexible method(model) can fit the data points better.

#### (b) The number of predictors p is extremely large, and the number of observations n is small.

Worse. The flexible method(model) would overfit the small number of observations.

#### (c) The relationship between the predictors and response is highly non-linear.

Better. The relationship is non-linear, which means we need a method(model) with high flexibility.

#### (d) The variance of the error terms, i.e. ^2 = Var(), is extremely high.

Worse. The flexible method(model) fits to the noise in the error terms and increases variance.
